```python
texts = df.apply(
    lambda row: f"Metric: {row['Metrics']}, Dec_11: {row['Dec 2011']}, Dec_12: {row['Dec 2012']}, Jun_22: {row['Jun 2022']}",
    axis=1
).tolist()
```

```python
formatted_texts = "\n".join(texts)
```

```python
formatted_texts
```

```python
from langchain_community.llms import Ollama

# Initialize the Ollama model
llm = Ollama(model="llama3.2:latest")

# Define the prompt template
prompt_template = "{data}\n\nWhat is the {metric} of {time_period}?, do not calculate"

# Fill in the dynamic parts
# prompt = prompt_template.format(data=formatted_texts, metric="equity capital", time_period="Jun-22")

prompt = f"""
For each of the following metrics, generate a question related to the metric and its values for Dec_11, Dec_12, and Jun_22. Then provide the answer to that question:

{formatted_texts}
"""

# Use the prompt in the model query
response = llm.invoke(input=prompt)
print(response)

```

```python
import json
qa_pairs = response.split('\n')

qa_json = []
for qa in qa_pairs:
    # Assuming each question-answer pair is separated by a ":"
    if ':' in qa:
        question, answer = qa.split(":", 1)
        qa_json.append({"question": question.strip(), "answer": answer.strip()})

# Save to a JSON file
with open("metrics_qa.json", "w") as json_file:
    json.dump(qa_json, json_file, indent=4)

print("JSON file with questions and answers saved.")
```

```python


# Open and load the JSON file
with open("metrics_qa.json", "r") as json_file:
    data = json.load(json_file)

# Print the contents of the JSON file
print(data)
```
