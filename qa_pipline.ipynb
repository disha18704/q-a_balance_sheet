```python
import pandas as pd
```

```python
df = pd.read_csv('data/balance_sheet_details.csv')
```

```python
df
```

```python
df['text'] = df.apply(lambda row: f"Metric: {row['Metrics']},Dec_11: {row['Dec 2011']},Dec_12: {row['Dec 2012']},Jun_22: {row['Jun 2022']}",axis=1)
```

```python
data=df.drop(columns=['Metrics','Dec 2011','Dec 2012','Jun 2022'])
```

```python
data['text'][0]
```

```python
texts = data['text'].tolist()
```

```python
texts
```

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import SentenceTransformerEmbeddings



embeddings_model = SentenceTransformerEmbeddings(model_name ="all-MiniLM-L6-v2")

vector_store = FAISS.from_texts(texts, embeddings_model)

vector_store.save_local("/Users/dishaarora/Desktop/q_a-balance_sheet/vector_store/balance_sheet_faiss_index")
```

```python
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline


model_name = "meta-llama/Llama-3B"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

llm_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=500)
llm = HuggingFacePipeline(pipeline=llm_pipeline)
```
